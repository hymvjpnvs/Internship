# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
#import library
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import requests
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
```

```python
#set up the Chrome driver and open the webpage
```

```python
driver = webdriver.Chrome()
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
#fetching required data
```

```python
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []
```

```python
rows = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr')

#scraping rank of the videos

try:
    for index, row in enumerate(rows[:30], start=1):
        Rank.append(str(index))
except NoSuchElementException:
    Rank.append("-")
except ElementNotVisibleException:
    Rank.append("--")
    
#scraping Name of the videos

name_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')

try:
    for i in name_tag[0:30]:
        v_name=i.text.split('[')
        Name.append(v_name[0])
except NoSuchElementException:
    Name.append("-")
except ElementNotVisibleException:
    Name.append("--")
    
#scraping Artist of the video

artist_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')

try:
    for i in artist_tag[0:30]:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
    
#scraping Upload date of the videos

date_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')

try:
    for i in date_tag[0:30]:
        Upload_date.append(i.text)
except NoSuchElementException:
    Upload_date.append("-")
    
#scraping Views of the videos

view_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')

try:
    for i in view_tag[0:30]:
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
```

```python
#closing webdriver
```

```python
driver.close()
```

```python
#creating DataFrame
```

```python
Top_videos = pd.DataFrame({})
Top_videos['Rank'] = Rank[:30]
Top_videos['Name'] = Name[:30]
Top_videos['Artist'] = Artist[:30]
Top_videos['Upload_date'] = Upload_date[:30]
Top_videos['Views(in Billions)'] = Views[:30]
Top_videos
```

# 4. Scrape the details of trending repositories on Github.com.

```python
#set up the Chrome driver and open the webpage
```

```python
driver = webdriver.Chrome()
url = 'https://github.com/'
driver.get(url)
```

```python
#click on Open Source

open_source = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]').click

#click on trending

trending = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/div/div[3]/ul/li[2]/a')
urls = trending.get_attribute("href")
driver.get(urls)
```

```python
#creating empty list
URLs = []
Repository_title = []
Description = []
Contributors = []
Language = []
```

```python
#scraping urls

repository = driver.find_elements(By.XPATH,'//a[@class="js-selected-navigation-item selected subnav-item"]')
for i in repository:
    URLs.append(i.get_attribute("href"))
    
#scraping Title data

title = driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]')
for i in title:
    Repository_title.append(i.text)
    
#scraping description data

description = driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in description:
    Description.append(i.text)
    
#scraping contributors data

contributors = driver.find_elements(By.XPATH,'//span[@class="d-inline-block mr-3"]')
for i in contributors:
    Contributors.append(i.text)
    
#scraping language data

language = driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
for i in language:
    Language.append(i.text)
```

```python
#creating DataFrame

Github = pd.DataFrame()
Github['Repository_title'] = Repository_title[:15]
Github['Repository_description'] = Description[:15]
Github['Contibutors_count'] = Contributors[:15]
Github['Language_used'] = Language[:15]


Github
```

# 8. Details of Datasets from UCI machine learning repositories

```python
#set up the Chrome driver and open the webpage
```

```python
driver = webdriver.Chrome()
url = 'https://archive.ics.uci.edu/'
driver.get(url)
```

```python
#click on all datasets 
all_dataset = driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
list_url = all_dataset.get_attribute("href")
driver.get(list_url)
time.sleep(4)
```

```python
#creating empty lists
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = [] 
Year = []
```

```python
expend = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]').click()
```

```python
#scraping data of dataset name
try:
    dataset = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in dataset:
        Dataset_name.append(i.text)
except NoSuchElementException:
    Dataset_name.append('-')
    
#scraping data of data type

try:
    data_type = driver.find_elements(By.XPATH,'//tbody[@class="border"]/tr/td[2]')
    for i in data_type:
        Data_type.append(i.text)
except NoSuchElementException:
    Data_type.append('-')

    
#scraping data of task

try:
    task = driver.find_elements(By.XPATH,'//div[1][@class="col-span-3 flex items-center gap-2"]//span[1]')
    for i in task:
        Task.append(i.text)
except NoSuchElementException:
    Task.append('-')
    
#scraping data of attribute type

try:
    attribute_type = driver.find_elements(By.XPATH,'//div[2][@class="col-span-3 flex items-center gap-2"]')
    for i in attribute_type:
        Attribute_type.append(i.text)
except NoSuchElementException:
    Attribute_type.append('-')
    
#scraping data of no of instances

try:
    instances = driver.find_elements(By.XPATH,'//div[3][@class="col-span-3 flex items-center gap-2"]//span[1]')
    for i in instances:
        No_of_instances.append(i.text)
except NoSuchElementException:
    No_of_instences.append('-')
    
#scraping data of no of attribute

try:
    attribute = driver.find_elements(By.XPATH,'//div[4][@class="col-span-3 flex items-center gap-2"]//span[1]')
    for i in attribute:
        No_of_attribute.append(i.text)
except NoSuchElementException:
    No_of_attribute.append('-') 
    
#scraping data of year

try:
    year = driver.find_elements(By.XPATH,'//tbody[@class="border"]/tr/td[3]')
    for i in year:
        Year.append(i.text[4:])
except NoSuchElementException:
    Year.append('-')
    time.sleep(3)
    
#creating DataFrame
Datasets = pd.DataFrame()
Datasets['Dataset Name'] = Dataset_name[:10]
Datasets['Data Type'] = Data_type[:10]
Datasets['Task'] = Task[:10]
Datasets['Attribute Type'] = Attribute_type[:10]
Datasets['No Of Instances'] = No_of_instances[:10]
Datasets['No Of Attribute'] = No_of_attribute[:10]
Datasets['Year'] = Year[:10]
Datasets
```

```python
#closing webdriver
driver.close()
```
