# 1. Write a python program to display IMDB’s Top rated 100 Indian movies’ data

```python
import pandas as pd
import requests
import re
import time
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
url = "https://www.imdb.com/list/ls056092300/"
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36'}
response = requests.get(url,headers = headers)
soup = BeautifulSoup(response.content, 'lxml')
```

```python
movie_title = soup.find_all('div', class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title")
movie_titles = []

for i in movie_title:
    movie_titles.append(i.a.h3.text)
movie_titles = [i.split('.')[1] for i in movie_titles]

print(movie_titles)
```

```python
rating = soup.find_all('span' , class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating")

movie_rating = []
for i in rating:
    movie_rating.append(i.text[0:3])

print(movie_rating)
```

```python
year = soup.find_all('span' , class_="sc-b189961a-8 kLaxqf dli-title-metadata-item")

movie_year = []
for i in year:
    year = i.text
    if re.match(r'^\d{4}$', year):
        movie_year.append(year)
        
print(movie_year)
```

```python
df = pd.DataFrame({
    'Name': movie_titles,
    'Rating': movie_rating,
    'Year of Release': movie_year
})
print(df)
```

# 2. Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the heading, date, content and the likes for the video from the link for the youtube video from the post.

```python
driver = webdriver.Chrome()
driver.get('https://www.patreon.com/coreyms')
time.sleep(10)
soup = BeautifulSoup(driver.page_source, 'html.parser')
posts = soup.find_all('div', class_='sc-xg7a7d-0 jJOVvL')
```

```python
headings = []
dates= []
contents = []
likes = []
video_links =[]
for post in posts:
    heading = post.find('span', class_ = 'sc-1cvoi1y-0 hxhWXn').text if post.find('span') else 'N/A'
    headings.append(heading)
    date = post.find('a', class_ = 'sc-kfPuZi doifNG').span.text if post.find('a') else 'N/A'
    dates.append(date)
    content = post.find('p').text if post.find('p') else 'N/A'
    contents.append(content)
    like = post.find('span', class_='sc-bqiRlB etUZPh').text if post.find('span', class_='sc-bqiRlB etUZPh') else 'N/A'
    likes.append(like)
    video_link = post.find('a', href=True)['href'] if post.find('a', href=True) else 'N/A'
    video_links.append(video_link)
driver.quit()
```

```python
for heading, date, content, like, video_link in zip(headings, dates, contents, likes, video_links):
    print(f"Heading: {heading}")
    print(f"Date: {date}")
    print(f"Content: {content}")
    print(f"Likes: {like}")
    print(f"Video Link: {video_link}")
    print("-" * 50)
```

# 4. Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/bestseller?sort=popular .

```python
url = 'https://www.bewakoof.com/bestseller?sort=popular.'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content)
products = soup.find_all('div', class_='plp-product-card')
```

```python
for i, product in enumerate(products[:10], 1):
    product_name = product.find('div', class_='productNaming bkf-ellipsis').h2.text.strip()
    price = product.find('div', class_='discountedPriceText clr-p-black false').text.strip()
    image_url = product.find('img')['src']
    
    print(f"Product {i}:")
    print(f"Name: {product_name}")
    print(f"Price: {price}")
    print(f"Image URL: {image_url}")
    print("\n")
```

# 5. Please visit https://www.cnbc.com/world/?region=world and scrap a) headings b) date c) News link

```python
url = 'https://www.cnbc.com/world/?region=world'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content)
articles_list = soup.find('ul', class_='LatestNews-list')
articles = articles_list.find_all('li')
print(articles)
```

```python
headings = []
news_dates = []
links = []
for article in articles:
        heading = article.find('div', class_='LatestNews-headlineWrapper').a.text
        headings.append(heading)
        date = article.find('time', class_ ="LatestNews-timestamp").text
        news_dates.append(date)
        link = article.find('a')['href']
        links.append(link)
```

```python
df = pd.DataFrame({
    'News Heading': headings,
    'Date': news_dates,
    'Link': links
})
print(df)
```

# 6. Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/ and scrap following details: Paper Title, Date, Author

```python
url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36'}
response = requests.get(url,headers = headers)
soup = BeautifulSoup(response.content, 'lxml')
```

```python
article_details = soup.find_all('div', class_ = 'article-listing')
print(article_details)
```

```python
titles = []
authors =[]
dates = []
for article in article_details:
    paper_title = article.find('h2', class_ = 'h5 article-title').a.text
    titles.append(paper_title)
    author_list = article.find('p', class_ = 'article-authors').text
    authors.append(author_list)
    published_date = article.find('p', class_ ='article-date').text
    dates.append(published_date)

authors = [re.sub(r'\|', ',', author) for author in authors]
```

```python
df = pd.DataFrame({
    'Paper Title': titles,
    'Authors': authors,
    'Published Date': dates
})
df = df.replace('\r\n','', regex=True)
print(df)
```
